```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv(r"E:\Courses\Data\files\Orders.csv",skiprows=4)
df
```

```python
df["Phone Number"]=df["Phone Number"].str.replace("Tel:","")
```

```python
df['City']=df['City'].str.title()
```

```python
df.info()
```

```python
df["Branch"]=df["Branch"].str.upper()
df["Email"]=df["Email"].str.lower()
```

```python
df
```

```python
df["Discount"]=df["Discount"].round(0)
# تتحكم هذه الدالة في عد الارقام العشرية 
```

```python
df
```

```python
df['Total Price']=df["Quantity"] * df["Unit Price"]
```

```python
df
```

```python
df["Discount"]=df["Discount"].fillna(0)
```

```python
df["Cost"]=df["Total Price"] - df["Discount"]
```

```python
df["Unit Price"]=df["Unit Price"].fillna(0)
```

```python
##اظهار كل الصفوف 
#pd.set_option("max_columns",None)
```

```python
df.insert(5,"Branch_code",new_column)
```

```python
df
```

```python
#filter
#loc[row,col]
df.loc[df["Status"]==True,"Status"]
df.loc[df["Status"]==True,"Order_Status"]="Done"
df.loc[(df["Status"]==False)&(df["Order Date"]<="1/1/2023"),"Order_Status"]="canceled"
```

```python
df
```

```python
def level(row):
    if row>1000 :
        return "High"
    else:
        return "Low"
    
df["Cost Level"]  =df["Cost"].apply(level)  
df
```

```python
df["Discount Discription"]=np.where(df["Discount"].isnull(),"No Discount",df["Discount"]/df["Cost"])
```
